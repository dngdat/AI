mtcars

```python
import pandas as pd
url = 'mtcars.csv'
data = pd.read_csv(url)

import matplotlib.pyplot as plt
data.hist() # hist số lần lặp lại 
plt.show
```

```python
correlation = data.corr()
print(correlation)
```

```python
import seaborn as sns
sns.heatmap(correlation, annot  = True)
```

```python
data.plot(kind = 'scatter', x = 'wt', y = 'mpg', figsize = (9, 9), color = 'black')
# kích thước 9 x 9 
```

```python
from sklearn import linear_model 
model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame(data['wt']), y = data['mpg']) #fit điều chỉnh 
print(model.intercept_)  
print(model.coef_)
# y = coef*x + intercept 
# coef: hệ số góc
model.score(X = pd.DataFrame(data['wt']), y = data['mpg']) # độ chính xác
```

```python
data.plot(kind = 'scatter', x = 'wt', y = 'mpg', figsize = (9, 9), color = 'black', xlim = (0, 7))
y_pred = model.predict(X = pd.DataFrame(data['wt']))
plt.plot(data['wt'], y_pred, color ='blue')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['wt'], data['wt']**2]).T
model.fit(X, y = data['mpg'])
print('Intercept của mô hình:', model.intercept_)
print('Coeffient của mô hình:', model.coef_)
print('Độ chính xác của mô hình:', model.score(X, y = data['mpg']))
```

```python
import numpy as np
data.plot(kind = 'scatter', x = 'wt', y = 'mpg', figsize = (9, 9), color = 'black', xlim = (0, 7))
range = np.arange(1.5, 5.5, 0.1)
predictor = pd.DataFrame([range, range**2]).T
y_pred = model.predict(X = predictor)
plt.plot(range, y_pred, color = 'blue')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['wt'], data['wt']**2, data['wt']**3]).T
model.fit(X, y = data['mpg'])
print('Intercept của mô hình:', model.intercept_)
print('Coeffient của mô hình:', model.coef_)
print('Độ chính xác của mô hình:', model.score(X, y = data['mpg']))
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['wt'], data['wt']**2, data['wt']**3, data['wt']**4]).T
model.fit(X, y = data['mpg'])
print('Intercept của mô hình:', model.intercept_)
print('Coeffient của mô hình:', model.coef_)
print('Độ chính xác của mô hình:', model.score(X, y = data['mpg']))

import numpy as np
data.plot(kind = 'scatter', x = 'wt', y = 'mpg', figsize = (9, 9), color = 'black', xlim = (0, 7))
range = np.arange(1.5, 5.5, 0.1)
predictor = pd.DataFrame([range, range**2, range**3, range**4]).T
y_pred = model.predict(X = predictor)
plt.plot(range, y_pred, color = 'blue')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['wt'], data['wt']**2, data['wt']**3, data['wt']**4, data['wt']**5]).T
model.fit(X, y = data['mpg'])
print('Intercept của mô hình:', model.intercept_)
print('Coeffient của mô hình:', model.coef_)
print('Độ chính xác của mô hình:', model.score(X, y = data['mpg']))

import numpy as np
data.plot(kind = 'scatter', x = 'wt', y = 'mpg', figsize = (9, 9), color = 'black', xlim = (0, 7))
range = np.arange(1.5, 5.5, 0.1)
predictor = pd.DataFrame([range, range**2, range**3, range**4, range **5]).T
y_pred = model.predict(X = predictor)
plt.plot(range, y_pred, color = 'blue')
```

```python
# hồi quy 2 biến 
model = linear_model.LinearRegression()
model.fit(X = data.loc[:, ['wt', 'hp']], y = data['mpg'])
print(model.intercept_)
print(model.coef_)
print(model.score(X = data.loc[:, ['wt', 'hp']], y = data['mpg']))

# mpg = -3.877wt - 0.031hp + 37.22
```

Salary 

```python
import pandas as pd
url = 'Salary_Data.csv'
data = pd.read_csv(url)

import matplotlib.pyplot as plt
data.hist() 
plt.show

correlation = data.corr()
print(correlation)
```

```python
import seaborn as sns # thư viện hiển thị dữ liệu
sns.pairplot(data)
```

```python
data.plot(kind = 'scatter', x = 'YearsExperience', y = 'Salary', figsize = (9, 9), color = 'black')
```

```python
x = data['YearsExperience']
y = data['Salary']
plt.plot(x, y, 'r')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
from sklearn import linear_model 
model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame(data['YearsExperience']), y = data['Salary'])
print('Intercept của mô hình:', model.intercept_)
print('Coeffient của mô hình:', model.coef_)
print('Độ chính xác của mô hình:', model.score(X = pd.DataFrame(data['YearsExperience']), y = data['Salary']))
```

```python
data.plot(kind = 'scatter', x = 'YearsExperience', y = 'Salary', figsize = (9, 9), color = 'black')
y_pred = model.predict(X = pd.DataFrame(data['YearsExperience']))
plt.plot(data['YearsExperience'], y_pred, color = 'blue')
```

```python
from sklearn import linear_model 
model = linear_model.LinearRegression()
X = pd.DataFrame([data['YearsExperience'], data['YearsExperience']**2, data['YearsExperience']**3,
                  data['YearsExperience']**4, data['YearsExperience']**5, data['YearsExperience']**6]).T
model.fit(X, y = data['Salary'])
print('Intercept của mô hình:', model.intercept_)
print('Coeffient của mô hình:', model.coef_)
print('Độ chính xác của mô hình:', model.score(X, y = data['Salary']))
```

```python
import numpy as np
data.plot(kind = 'scatter', x = 'YearsExperience', y = 'Salary', figsize = (9, 9), color = 'black')
range = np.arange(data['YearsExperience'].min(), data['YearsExperience'].max(), 0.1)
predictor = pd.DataFrame([range, range**2, range**3, range**4, range**5, range**6]).T
y_pred = model.predict(X = predictor)
plt.plot(range, y_pred, color = 'blue')
```
